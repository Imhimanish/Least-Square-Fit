## **Least Square Fit**

```python
import numpy as np
import matplotlib.pyplot as plt
X = np.array([1,2,3,4,5])
Y = np.array([3.1,5.9,7.2,10.0,11.3])
n = X.size
sx = np.sum(X)
sy = np.sum(Y)
sx2 = np.sum(X**2)
sxy = np.sum(X*Y)
a1 = (sxy-sx*sy)/(sx2-sx**2)
a0 = sy-a1*sx
Yfit = a0 + a1*X
plt.plot(X,Y,'o')
plt.plot(X,Yfit)
plt.show()
```
